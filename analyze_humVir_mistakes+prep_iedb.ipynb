* Add extra features from SB
For human proteins:
*  endogenous retroviruses. Shorter length (< 170). Poorly characterized proteins (with less than 8 annotation keywords) and Protein names containing "putative". "Reference proteome" designated proteins were also more likely to be mischaracterized. 


````python
df["Endogenous Retrovirus"] = (df["Protein names"].str.contains("endogenous",case=False,na=False)) & (df["virus"]==0)
df["Short sequence < 170"] = df["Length"]<170
df["Putative"] = df["Protein names"].str.contains("putativ",case=False,na=False)
df["Few Keywords < 8"] = df["Keywords"].str.split().str.len()<8
````

````python
hum_seg_cols = ['Endogenous Retrovirus', 'Short sequence < 170',
       'Putative', 'Few Keywords < 8']

BASELINE_HUM_ERR_RATE = df[df["virus"]==0]["mistake"].mean()
print("BASELINE_HUM_ERR_RATE",BASELINE_HUM_ERR_RATE)
BASE_ERR_RATE =  df["mistake"].mean()
print("BASE_ERR_RATE",BASE_ERR_RATE)
for c in hum_seg_cols:
    print(c)
    ## .dropna(subset=c,axis=0)
    grouped = df[[c,"mistake"]].groupby(c)
    # grouped = grouped.filter(lambda x: len(x) > 4)  ## Doesn't work when here
    res = grouped["mistake"].agg(["mean","size","sum"]).round(4) ## size is same as count # ,"count"
    res["lift"] = res["mean"].div(BASE_ERR_RATE).round(2)
    # res = res[res["count"]>= 7].sort_values(["mean","count","lift"],ascending=False)
    res["mean"] = (100*res["mean"]).round(3)
    display(res)
    print()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def make_iedb_4group_plots(df):
    # Determine the global min and max of 'IEDB_immunogenicity_score' for consistent x-axis limits
    min_score = df['IEDB_immunogenicity_score'].min()
    max_score = df['IEDB_immunogenicity_score'].max()
    
    # Create a 2x2 subplot with shared x-axis limits
    fig, axs = plt.subplots(2, 2, figsize=(12, 10), sharex=True)
    
    # Define the sub-dataframes for each condition
    data_true_virus1 = df[(df['mistake'] == True) & (df['virus'] == 1)]
    data_true_virus0 = df[(df['mistake'] == True) & (df['virus'] == 0)]
    data_false_virus1 = df[(df['mistake'] == False) & (df['virus'] == 1)]
    data_false_virus0 = df[(df['mistake'] == False) & (df['virus'] == 0)]
    
    # A helper function to plot the mean and median on the axes
    def add_stats_text(ax, data):
        mean_val = data['IEDB_immunogenicity_score'].mean()
        median_val = data['IEDB_immunogenicity_score'].median()
        ax.text(0.05, 0.95, f'Mean: {mean_val:.2f}\nMedian: {median_val:.2f}',
                transform=ax.transAxes, fontsize=14,
                verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))
    
    # Plot each condition with statistics
    sns.histplot(data=data_true_virus1, x='IEDB_immunogenicity_score', color='blue', kde=True, ax=axs[0, 0], binwidth=0.5)
    add_stats_text(axs[0, 0], data_true_virus1)
    axs[0, 0].set_title('Virus, Mistake = True')
    axs[0, 0].set_xlim(min_score, max_score)
    
    sns.histplot(data=data_true_virus0, x='IEDB_immunogenicity_score', color='red', kde=True, ax=axs[0, 1], binwidth=0.5)
    add_stats_text(axs[0, 1], data_true_virus0)
    axs[0, 1].set_title('Human,Mistake = True')
    axs[0, 1].set_xlim(min_score, max_score)
    
    sns.histplot(data=data_false_virus1, x='IEDB_immunogenicity_score', color='green', kde=True, ax=axs[1, 0], binwidth=0.5)
    add_stats_text(axs[1, 0], data_false_virus1)
    axs[1, 0].set_title('Virus,Mistake = False')
    axs[1, 0].set_xlim(min_score, max_score)
    
    sns.histplot(data=data_false_virus0, x='IEDB_immunogenicity_score', color='orange', kde=True, ax=axs[1, 1], binwidth=0.5)
    add_stats_text(axs[1, 1], data_false_virus0)
    axs[1, 1].set_title('Human, Mistake = False')
    axs[1, 1].set_xlim(min_score, max_score)
    
    plt.tight_layout()
    plt.show()

# Example usage:
## Immunogenicity Score 
make_iedb_4group_plots(df_iedb_subset) 
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def make_iedb_4group_plots(df):
    # Determine the global min and max of 'Immunogenicity score' for consistent x-axis limits
    min_score = df['IEDB_immunogenicity_score'].min()
    max_score = df['IEDB_immunogenicity_score'].max()
    
    # Create a 2x2 subplot with shared x-axis limits and higher DPI
    fig, axs = plt.subplots(2, 2, figsize=(12, 10), sharex=True, dpi=600)
    
    # Define the sub-dataframes for each condition
    data_groups = [
        (df[(df['mistake'] == True) & (df['virus'] == 1)], 'Virus, Mistake = True', 'blue'),
        (df[(df['mistake'] == True) & (df['virus'] == 0)], 'Human, Mistake = True', 'red'),
        (df[(df['mistake'] == False) & (df['virus'] == 1)], 'Virus, Mistake = False', 'green'),
        (df[(df['mistake'] == False) & (df['virus'] == 0)], 'Human, Mistake = False', 'orange')
    ]
    
    for idx, (data, title, color) in enumerate(data_groups):
        ax = axs[idx // 2, idx % 2]
        sns.histplot(data=data, x='IEDB_immunogenicity_score', color=color, kde=True, ax=ax, binwidth=0.5)
        ax.set_title(title, fontsize=16, pad=10)
        ax.set_xlim(min_score, max_score)
        ax.set_xlabel('Immunogenicity score')
        ax.grid(True, which='major', linestyle='--', linewidth='0.5', color='grey')
        
        # Annotation for mean and median
        mean_val = data['IEDB_immunogenicity_score'].mean()
        median_val = data['IEDB_immunogenicity_score'].median()
        ax.text(0.05, 0.95, f'Mean: {mean_val:.2f}\nMedian: {median_val:.2f}',
                transform=ax.transAxes, fontsize=12,
                verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))
        
        sns.despine(ax=ax)
    
    # Set common labels
    for ax in axs.flat:
        ax.set(xlabel='Immunogenicity score')
        ax.label_outer()  # Hide x labels and tick labels for top plots and y ticks for right plots.
    
    # Improve the layout and show the plot
    plt.subplots_adjust(hspace=0.3, wspace=0.2)
    plt.show()

# Example usage:
# Replace 'your_dataframe_here' with your actual DataFrame variable name.
make_iedb_4group_plots(df_iedb_subset)
````
